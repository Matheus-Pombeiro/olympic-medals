# **Olympic Medals**

## NumPy and Matplotlib

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Getting the dataset url and adding it to the project's dataset
url = 'https://raw.githubusercontent.com/Matheus-Pombeiro/olympic-medals/main/olympics2024.csv'
medals = np.loadtxt(url, delimiter=',', usecols=np.arange(3, 7, 1), skiprows=1)
medals
```

Exploring the dataset

```python
# Finding the number of dimensions (lines and cols)
dimensions = medals.ndim
# Finding the number of elements inside the dataset
elements = medals.size
# Finding the number of elements in each dimension (lines and cols)
dimensions_elements = medals.shape
# Presenting these informations
print(f'Medals dataset - Information: {dimensions} dimension(s), {elements} element(s) and {dimensions_elements} element(s) / dimension(s)')
```

```python
# Making a data transposition on the dataset
transposed_medals = medals.T
transposed_medals
```

```python
# Getting the gold medals
gold_medals = transposed_medals[0, :]
# Getting the silver medals
silver_medals = transposed_medals[1, :]
# Getting the bronze medals
bronze_medals = transposed_medals[2, :]
# Getting the total of medals
total_medals = transposed_medals[3, :]
```

Processing the gold medals

```python
# Calculating the average
gold_medals_mean = np.mean(gold_medals)
# Calculating the median
gold_medals_median = np.median(gold_medals)
# Calculating the standard deviation
gold_medals_std = np.std(gold_medals)
# Presenting the results
print(f'Gold Medals - Information: mean: {gold_medals_mean:.1f}, median: {gold_medals_median:.1f} and standard devidation: {gold_medals_std:.1f}')
```

```python
# Plotting a graph with the number of gold medals and the total of countries
plt.bar(gold_medals, np.arange(0, len(gold_medals), 1), color='gold')
plt.title('Gold medals - Countries')
plt.legend(['Medals earned'])
plt.ylabel('Countries')
plt.xlabel('Gold Medals')
```

Processing the silver medals

```python
# Calculating the average
silver_medals_mean = np.mean(silver_medals)
# Calculating the median
silver_medals_median = np.median(silver_medals)
# Calculating the standard deviation
silver_medals_std = np.std(silver_medals)
# Presenting the results
print(f'Silver Medals - Information: mean: {silver_medals_mean:.1f}, median: {silver_medals_median:.1f} and standard devidation: {silver_medals_std:.1f}')
```

```python
# Plotting a graph with the number of silver medals and the total of countries
plt.bar(silver_medals, np.arange(0, len(silver_medals), 1), color='silver')
plt.title('Silver medals - Countries')
plt.legend(['Medals earned'])
plt.ylabel('Countries')
plt.xlabel('Silver Medals')
```

Processing the bronze medals

```python
# Calculating the average
bronze_medals_mean = np.mean(bronze_medals)
# Calculating the median
bronze_medals_median = np.median(bronze_medals)
# Calculating the standard deviation
bronze_medals_std = np.std(bronze_medals)
# Presenting the results
print(f'Bronze Medals - Information: mean: {bronze_medals_mean:.1f}, median: {bronze_medals_median:.1f} and standard devidation: {bronze_medals_std:.1f}')
```

```python
# Plotting a graph with the number of bronze medals and the total of countries
plt.bar(bronze_medals, np.arange(0, len(bronze_medals), 1), color='brown')
plt.title('Bronze medals - Countries')
plt.legend(['Medals earned'])
plt.ylabel('Countries')
plt.xlabel('Bonze Medals')
```

Processing the total of medals

```python
# Calculating the average
total_medals_mean = np.mean(total_medals)
# Calculating the median
total_medals_median = np.median(total_medals)
# Calculating the standard deviation
total_medals_std = np.std(total_medals)
# Presenting the results
print(f'Total of Medals - Information: mean: {total_medals_mean:.1f}, median: {total_medals_median:.1f} and standard devidation: {total_medals_std:.1f}')
```

```python
# Plotting a graph with the number of total of medals and the total of countries
plt.bar(total_medals, np.arange(0, len(total_medals), 1), width=1, color='blue')
plt.title('Total of medals - Countries')
plt.legend(['Medals earned'])
plt.ylabel('Countries')
plt.xlabel('Total of Medals')
```

```python
# Adding up the number of each kind of medal
gold_medals_sum = np.sum(gold_medals)
silver_medals_sum = np.sum(silver_medals)
bronze_medals_sum = np.sum(bronze_medals)
# Plotting a graph containing the percentage of each kind of medal
plt.pie([gold_medals_sum, silver_medals_sum, bronze_medals_sum], labels=['Gold', 'Silver', 'Bronze'], autopct='%1.1f%%', startangle=90, colors=['gold', 'silver', 'brown'])
plt.title('Total of Medals (%)')
```

Processing the total of medals of the first place

```python
# Getting the number of each kind of medal of the first place
first_gold_medals = gold_medals[0]
first_silver_medals = silver_medals[0]
first_bronze_medals = bronze_medals[0]
# Presenting the number of each kind of medals earned by the first place
print(f'Medals: Gold: {first_gold_medals}, Silver: {first_silver_medals}, Bronze: {first_bronze_medals}')
```

```python
# Plotting a graph containing the percentage of each kind of medal of the first place
plt.pie([first_gold_medals, first_silver_medals, first_bronze_medals], labels=['Gold', 'Silver', 'Bronze'], autopct='%1.1f%%', colors=['gold', 'silver', 'brown'], startangle=90)
plt.title('First Place Medals')
```

Processing the number of medals of the last place

```python
# Getting the number of each kind of medal of the last place
last_gold_medals = gold_medals[-1]
last_silver_medals = silver_medals[-1]
last_bronze_medals = bronze_medals[-1]
# Presenting the number of each kind of medals earned by the last place
print(f'Medals: Gold: {last_gold_medals}, Silver: {last_silver_medals}, Bronze: {last_bronze_medals}')
```

```python
# Plotting a graph containing the percentage of each kind of medal of the last place
plt.pie([last_bronze_medals], labels=['Bronze'], autopct='%1.1f%%', colors=['brown'], startangle=90)
plt.title('Last Place Medals')
```
