# **Olympic Medals**

## NumPy and Matplotlib

Importing NumPy and Matplotlib and getting the dataset

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Getting the dataset url and adding it to the project's dataset
url = 'https://raw.githubusercontent.com/Matheus-Pombeiro/olympic-medals/main/olympics2024.csv'
medals = np.loadtxt(url, delimiter=',', usecols=np.arange(3, 7, 1), skiprows=1)
medals
```

Exploring the dataset

```python
# Finding the number of dimensions (lines and cols)
dimensions = medals.ndim
# Finding the number of elements inside the dataset
elements = medals.size
# Finding the number of elements in each dimension (lines and cols)
dimensions_elements = medals.shape
# Presenting these informations
print(f'Medals dataset - Information: {dimensions} dimension(s), {elements} element(s) and {dimensions_elements} element(s) / dimension(s)')
```

Preparing the dataset

```python
# Making a data transposition on the dataset
transposed_medals = medals.T
transposed_medals
```

```python
# Getting the gold medals
gold_medals = transposed_medals[0, :]
# Getting the silver medals
silver_medals = transposed_medals[1, :]
# Getting the bronze medals
bronze_medals = transposed_medals[2, :]
# Getting the total medals
total_medals = transposed_medals[3, :]
```

Processing the gold medals

```python
# Calculating the average
gold_medals_mean = np.mean(gold_medals)
# Calculating the median
gold_medals_median = np.median(gold_medals)
# Calculating the standard deviation
gold_medals_std = np.std(gold_medals)
# Presenting the results
print(f'Gold Medals - Information: mean: {gold_medals_mean:.1f}, median: {gold_medals_median:.1f} and standard devidation: {gold_medals_std:.1f}')
```

```python
# Plotting a graph with the number of gold medals and the total countries
plt.bar(gold_medals, np.arange(0, len(gold_medals), 1), color='gold')
plt.title('Gold medals - Countries')
plt.legend(['Medals won'])
plt.ylabel('Countries')
plt.xlabel('Gold Medals')
```

Processing the silver medals

```python
# Calculating the average
silver_medals_mean = np.mean(silver_medals)
# Calculating the median
silver_medals_median = np.median(silver_medals)
# Calculating the standard deviation
silver_medals_std = np.std(silver_medals)
# Presenting the results
print(f'Silver Medals - Information: mean: {silver_medals_mean:.1f}, median: {silver_medals_median:.1f} and standard devidation: {silver_medals_std:.1f}')
```

```python
# Plotting a graph with the number of silver medals and the total countries
plt.bar(silver_medals, np.arange(0, len(silver_medals), 1), color='silver')
plt.title('Silver medals - Countries')
plt.legend(['Medals won'])
plt.ylabel('Countries')
plt.xlabel('Silver Medals')
```

Processing the bronze medals

```python
# Calculating the average
bronze_medals_mean = np.mean(bronze_medals)
# Calculating the median
bronze_medals_median = np.median(bronze_medals)
# Calculating the standard deviation
bronze_medals_std = np.std(bronze_medals)
# Presenting the results
print(f'Bronze Medals - Information: mean: {bronze_medals_mean:.1f}, median: {bronze_medals_median:.1f} and standard devidation: {bronze_medals_std:.1f}')
```

```python
# Plotting a graph with the number of bronze medals and the total countries
plt.bar(bronze_medals, np.arange(0, len(bronze_medals), 1), color='brown')
plt.title('Bronze medals - Countries')
plt.legend(['Medals won'])
plt.ylabel('Countries')
plt.xlabel('Bonze Medals')
```

Processing the total medals

```python
# Calculating the average
total_medals_mean = np.mean(total_medals)
# Calculating the median
total_medals_median = np.median(total_medals)
# Calculating the standard deviation
total_medals_std = np.std(total_medals)
# Presenting the results
print(f'Total Medals - Information: mean: {total_medals_mean:.1f}, median: {total_medals_median:.1f} and standard devidation: {total_medals_std:.1f}')
```

```python
# Plotting a graph with the number total medals and the total countries
plt.bar(total_medals, np.arange(0, len(total_medals), 1), width=1, color='blue')
plt.title('Total medals - Countries')
plt.legend(['Medals won'])
plt.ylabel('Countries')
plt.xlabel('Total Medals')
```

```python
# Adding up the number of each kind of medal
gold_medals_sum = np.sum(gold_medals)
silver_medals_sum = np.sum(silver_medals)
bronze_medals_sum = np.sum(bronze_medals)
# Plotting a graph containing the percentage of each kind of medal
plt.pie([gold_medals_sum, silver_medals_sum, bronze_medals_sum], labels=['Gold', 'Silver', 'Bronze'], autopct='%1.1f%%', startangle=90, colors=['gold', 'silver', 'brown'])
plt.title('Total Medals (%)')
```

Processing the total medals of the first place

```python
# Getting the number of each kind of medal of the first place
first_gold_medals = gold_medals[0]
first_silver_medals = silver_medals[0]
first_bronze_medals = bronze_medals[0]
first_total_medals = total_medals[0]
# Presenting the number of each kind of medal won by the first place
print(f'Medals: Gold: {first_gold_medals}, Silver: {first_silver_medals}, Bronze: {first_bronze_medals}, Total: {first_total_medals}')
```

```python
# Plotting a graph containing the percentage of each kind of medal of the first place
plt.pie([first_gold_medals, first_silver_medals, first_bronze_medals], labels=['Gold', 'Silver', 'Bronze'], autopct='%1.1f%%', colors=['gold', 'silver', 'brown'], startangle=90)
plt.title('First Place Medals')
```

Processing the number of medals of the last place

```python
# Getting the number of each kind of medal of the last place
last_gold_medals = gold_medals[-1]
last_silver_medals = silver_medals[-1]
last_bronze_medals = bronze_medals[-1]
last_total_medals = total_medals[-1]
# Presenting the number of each kind of medal won by the last place
print(f'Medals: Gold: {last_gold_medals}, Silver: {last_silver_medals}, Bronze: {last_bronze_medals}, Total: {last_total_medals}')
```

```python
# Plotting a graph containing the percentage of each kind of medal of the last place
plt.pie([last_bronze_medals], labels=['Bronze'], autopct='%1.1f%%', colors=['brown'], startangle=90)
plt.title('Last Place Medals')
```

## Pandas

Importing Pandas and getting the dataset

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/Matheus-Pombeiro/olympic-medals/main/olympics2024.csv'
medals = pd.read_csv(url, sep=',')
```

Extracting some information from the dataset

```python
# The top ten countries
medals.head(10)
```

```python
# Lines x columns
medals.shape
```

```python
# List of columns
medals.columns
```

```python
# Information about each column
medals.info()
```

First placed countries

```python
# Getting the five first placed
first_placed = medals.head(5)
first_placed
```

```python
# Plotting the first placed countries by their medals (gold, silver and bronze)
first_placed_medals = first_placed.groupby(['Country', 'Rank'])[['Bronze', 'Silver', 'Gold']].sum().sort_values('Gold')
first_placed_medals.plot(kind='barh', figsize=(12, 8), color=['brown', 'silver', 'gold'], title='First Placed Countries - Medals', xlabel='Number of Medals')
```

```python
# Plotting the total medals won by the first placed countries
first_placed_total_medals = first_placed.groupby(['Country', 'Rank'])[['Total']].sum().sort_values('Total')
first_placed_total_medals.plot(kind='barh', figsize=(12, 6), title='First Placed Countries - Total Medals', xlabel='Number of Medals')
```

```python
# Plotting the percentage of total medals won by the first placed countries
first_placed_total_medals.plot(kind='pie', figsize=(9, 9), subplots=True, autopct='%1.1f%%', startangle=90, title='First Placed Countries - Total Medals (%)')
```

Finding the better results by silver medals

```python
# Finding countries that won more than 20 silver medals
better_silver_medals = medals.query('Silver >= 20')[['Rank', 'Country', 'Silver']]
better_silver_medals.head()
```

```python
# Plotting the countries that won more than 20 silver medals
better_silver_medals = better_silver_medals.groupby(['Country', 'Rank'])[['Silver']].sum().sort_values('Silver')
better_silver_medals.plot(kind='barh', figsize=(12, 6), color='silver', title='Best Achievements by Silver Medals', xlabel='Number of Medals')
```

Finding the better results by bronze medals

```python
# Finding countries that won more than 20 bronze medals
better_bronze_medals = medals.query('Bronze >= 20')[['Rank', 'Country', 'Bronze']]
better_bronze_medals.head()
```

```python
# Plotting the countries that won more than 20 bronze medals
better_bronze_medals = better_bronze_medals.groupby(['Country', 'Rank'])[['Bronze']].sum().sort_values('Bronze')
better_bronze_medals.plot(kind='barh', figsize=(12, 6), color='brown', title='Best Achievements by Bronze Medals', xlabel='Number of Medals')
```

Finding the better results by total medals

```python
# Finding countries that won more than 50 total medals
better_total_medals = medals.query('Total >= 50')[['Rank', 'Country', 'Total']]
better_total_medals.head()
```

```python
# Plotting the countries that won more than 50 total medals
better_total_medals = better_total_medals.groupby(['Country', 'Rank'])[['Total']].sum().sort_values('Total')
better_total_medals.plot(kind='barh', figsize=(12, 6), color='blue', title='Best Achievements by Total Medals', xlabel='Number of Medals')
```

```python
# Plotting the percentage of the countries that won more than 50 total medals
better_total_medals.plot(kind='pie', figsize=(9, 9), subplots=True, autopct='%1.1f%%', startangle=90, title='Best Achievements by Total Medals (%)')
```
